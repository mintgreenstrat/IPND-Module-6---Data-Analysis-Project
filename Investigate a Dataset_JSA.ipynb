## Section 3: Exploration

### 1. Which team has the most wins? Which team scored the most goals?

There are two questions here. I'll designate them 1a and 1b and work them separately. Both questions can be answered by using a subset of the data in match_df and adding a couple of new columns. 

#### 1a. Which team won the most games?
Step 1: I only need a the first few columns of *match_df* in order to answer this so I'll exclude all the columns to the right of [away_team_goal] using .loc

```python
match_df_slice = match_df.loc[:, :'away_team_goal']
check_dtypes(match_df_slice)
```

Step 2: I'm only interested in the Bundesliga so I'll filter the slice for [league id] = 7809 and then check the output using .describe()

```python
match_df_slice_germany = match_df_slice.loc[match_df_slice['league_id']==7809]
match_df_slice_germany.describe()

# comparing the number of games here (2448) with using the same league_id filter in DB Browser gives me confidence :-)
```

Step 3: In order to count up the number of wins for each team I need a column that will be used to exclude all games that are a Draw. I'll write a simple function to use the condition row['home_team_goal'] != row['away_team_goal'] and .apply() this to a new column.

```python
def win(row):
    if row['home_team_goal'] != row['away_team_goal']:
        return 1
    else:
        return 0

match_df_slice_germany['win'] = match_df_slice_germany.apply(win, axis = 1)
match_df_slice_germany

# I acknowledge the warning that is raised below but reading through the suggestion I don't see how I would do this with .loc
```

Step 4: In order to sum up the number of wins for each team, I need to know which team wins each game. I created another column using another simple function as below.

```python
def win_team(row):
    if (row['win'] == 1) and (row['home_team_goal'] > row['away_team_goal']):
        return row['home_team_api_id']
    elif (row['win'] == 1) and (row['away_team_goal'] > row['home_team_goal']):
        return row['away_team_api_id']
    else:
        return 0

match_df_slice_germany['win_team'] = match_df_slice_germany.apply(win_team, axis = 1)
match_df_slice_germany

# I acknowledge the warning that is raised below but reading through the suggestion I don't see how I would do this with .loc given I have three conditions...?
```

Step 5: I cannot .groupby() using the [home_team_api_id] or [away_team_api_id] as these are int data types. I need to replace the team ids with their [team_long_name] in the team_df. I have done this using .map() which I reference a procedure in the cell below.

```python
match_df_slice_germany['win_team'] = match_df_slice_germany['win_team'].map(team_df.set_index('team_api_id')['team_long_name'])
match_df_slice_germany

# adapated from ref[3]
```

Step 6: In order to end up with the data in the format I wanted I have taken a number of sequential steps. 
 - slice again for only the columns needed in the answer
 - rearrange the column order
 - remove rows that are draws (win = 0)
 - groupby [win_team] and sum up the [win] column
 - convert to pd.Series
 - sort values into descending order

```python
answer1a = match_df_slice_germany.loc[:, 'win':'win_team']
answer1a = answer1a[['win_team','win']]
answer1a = answer1a[answer1a.win != 0]
answer1a = answer1a.groupby(['win_team'])['win'].agg('sum')
answer1a = pd.Series(answer1a)
answer1a = answer1a.sort_values(ascending=False)
answer1a

# adapted the groupby step from ref[4]
```

Step 7: Explore the data and plot it

```python
answer1a.describe()
```

```python
def stan_column(column):
    return (column - column.mean()) / column.std(ddof=0)

def standardize(df):
    return df.apply(stan_column)

answer1a_df = pd.DataFrame(answer1a)
answer1a_stan = standardize(answer1a_df)
answer1a_stan
```

```python
df_1a = answer1a_df.merge(answer1a_stan, on='win_team', how='inner')
df_1a = df_1a.rename(columns={'win_x': 'Total Wins', 'win_y': 'Standardized Wins'})
df_1a
```

```python
output = answer1a.plot(kind='bar')
output.set_ylabel('Total Wins')
output.set_xlabel('Team')
output.set_title('Bundesliga 2008/09 through 2015/16 Seasons')
```

```python
output10 = answer1a_stan.plot(kind='bar')
output10.set_ylabel('Standardized Wins')
output10.set_xlabel('Team')
output10.set_title('Bundesliga 2008/09 through 2015/16 Seasons')
output10.legend().remove()
```

#### Conclusions 1a) Which team has the most wins?
 - The team with the most wins is FC Bayern Munich with 193 wins over the time period investigated
 - 193 wins is 2.5 standard deviations above the mean over the time period investigated
 - FC Bayern Munich also has a big gap of 0.7 standard deviations from it's nearest competitor over the time period investigated

#### 1b. Which team scored the most goals?
I can use the same slice of match_df even though I don't need the [win] or [win_team] columns. 

Step 1: As above (1a, Step 5) I'll use the same procedure to map in the [team_long_name] for [home_team_api_id] and [away_team_api_id] given I will be using these in groupby() in the next steps.

```python
match_df_slice_germany['home_team_api_id'] = match_df_slice_germany['home_team_api_id'].map(team_df.set_index('team_api_id')['team_long_name'])
match_df_slice_germany['away_team_api_id'] = match_df_slice_germany['away_team_api_id'].map(team_df.set_index('team_api_id')['team_long_name'])
match_df_slice_germany
```

```python
answer1b_home = match_df_slice_germany.loc[:, ['home_team_api_id', 'home_team_goal']]
answer1b_home = answer1b_home.groupby(['home_team_api_id'])['home_team_goal'].agg('sum')
answer1b_home
```

Step 3: I'll repeat the same step for the away games and goals. 

```python
answer1b_away = match_df_slice_germany.loc[:, ['away_team_api_id', 'away_team_goal']]
answer1b_away = answer1b_away.groupby(['away_team_api_id'])['away_team_goal'].agg('sum')
answer1b_away
```

Step 4: Similarly to above (1a, Step 6) I'll bring together the home and away goals for each team by adding the two series together, convert to a pd.Series and then sort.

```python
answer1b_total = answer1b_home + answer1b_away
answer1b_total = pd.Series(answer1b_total)
answer1b_total = answer1b_total.sort_values(ascending=False)
answer1b_total
```

### Question 2: Most improved team in the Bundesliga by wins and goals scored?
Again, this is a two part question so I'll break out each as separate questions and bring together the results in a conclusions section.

#### Question 2a: Most improved team by total wins?
In order to answer this question, I can again make use of the same slice of the match_df. However, this time I will need to compare the total wins in 2008/09 season against the total wins in the 2015/16 season. The biggest difference is defined as the most improved.

Step 1: Take a new slice of the match_df for only the columns I need ([season], [home_team_api_id], [win]). Then I need too;
 - remove all draws
 - filter for season 2008/2009
 - groupby [home_team_api_id] and sum win column
 - convert to pd.DataFrame

```python
answer2a_home_2008 = match_df_slice_germany.loc[:, ['season', 'home_team_api_id', 'win']]
answer2a_home_2008 = answer2a_home_2008[answer2a_home_2008.win != 0]
answer2a_home_2008 = answer2a_home_2008[answer2a_home_2008.season == '2008/2009']
answer2a_home_2008 = answer2a_home_2008.groupby(['home_team_api_id'])['win'].agg('sum')
answer2a_home_2008 = pd.DataFrame(data = answer2a_home_2008)
answer2a_home_2008
```

Step 2: Repeat for the away games in 2008/2009 season

```python
answer2a_away_2008 = match_df_slice_germany.loc[:, ['season', 'away_team_api_id', 'win']]
answer2a_away_2008 = answer2a_away_2008[answer2a_away_2008.win != 0]
answer2a_away_2008 = answer2a_away_2008[answer2a_away_2008.season == '2008/2009']
answer2a_away_2008 = answer2a_away_2008.groupby(['away_team_api_id'])['win'].agg('sum')
answer2a_away_2008 = pd.DataFrame(data = answer2a_away_2008)
answer2a_away_2008
```

Step 3: Combine dataframes for total wins by team in 2008/2009 season

```python
answer2a_total_2008 = answer2a_home_2008 + answer2a_away_2008
answer2a_total_2008 = answer2a_total_2008.sort_values(by='win', ascending=False)
answer2a_total_2008
```

Step 4: Repeat steps 1 - 3 for the 2015/2016 season

```python
#Step 1
answer2a_home_2015 = match_df_slice_germany.loc[:, ['season', 'home_team_api_id', 'win']]
answer2a_home_2015 = answer2a_home_2015[answer2a_home_2015.win != 0]
answer2a_home_2015 = answer2a_home_2015[answer2a_home_2015.season == '2015/2016']
answer2a_home_2015 = answer2a_home_2015.groupby(['home_team_api_id'])['win'].agg('sum')
answer2a_home_2015 = pd.DataFrame(data = answer2a_home_2015)
answer2a_home_2015

#Step 2
answer2a_away_2015 = match_df_slice_germany.loc[:, ['season', 'away_team_api_id', 'win']]
answer2a_away_2015 = answer2a_away_2015[answer2a_away_2015.win != 0]
answer2a_away_2015 = answer2a_away_2015[answer2a_away_2015.season == '2015/2016']
answer2a_away_2015 = answer2a_away_2015.groupby(['away_team_api_id'])['win'].agg('sum')
answer2a_away_2015 = pd.DataFrame(data = answer2a_away_2015)
answer2a_away_2015

#Step 3
answer2a_total_2015 = answer2a_home_2015 + answer2a_away_2015
answer2a_total_2015 = answer2a_total_2015.sort_values(by='win', ascending=False)
answer2a_total_2015
```

Step 5: Join the two tables to compare the values for 2008/09 and 2015/16 seasons with an inner join as the same teams were not playing in the league in 2008/09 and 2015/16. Otherwise there would be NaN in a bunch of the columns. 

```python
answer2a = pd.merge(answer2a_total_2008, answer2a_total_2015, on='home_team_api_id', how='inner')
answer2a = answer2a.rename(columns={'home_team_api_id' : 'team_long_name', 'win_x' : '2008/09', 'win_y' : '2015/16'})
answer2a
```

Step 6: Add a column to calculate the % change from 2008/09 to 2015/16 seasons. Then sort by [% Change]

```python
def change(row):
    return ((row['2015/16'] - row['2008/09']) / row['2008/09']) * 100
    
answer2a['% Change'] = answer2a.apply(change, axis=1)
answer2a = answer2a.sort_values(by='% Change', ascending=False)
answer2a
```

```python
answer2a.describe()
```

```python
output3 = answer2a.plot(kind='bar')
output3.set_ylabel('Team Wins')
output3.set_xlabel('Team')
output3.set_title('Bundesliga 2015/16 vs 2008/09 Seasons')
```

 - FC Bayern Munich is the stand out team from both a total wins and goals scored perspective over the time period investigated.

#### Question 2: Conclusions

 - The most improved team by total wins is Borussia Dortmund with an increase of 8 wins in their 2015/16 season vs 2008/09. 
 - The most improved team by goals scored is Borussia Monchengladbach with an increase of 28 additional goals scored in their 2015/16 season vs 2008/09.

#### Overall: Limitations
 - There are many limitations to this dataset and analysis not limited to the ones made explicit here in the Conclusions or as notes in the Markdown commentary throughout. Here are the three most important from the authors perspective:
    * The dataset covers only seasons 2008/09 through 2015/16. The analysis makes no claims for trends or conclusions drawn across earlier or later seasons.

    * Question 3 suffers from a severe lack of data points in order to draw statistically signficant conclusions. These answers should not be viewed as more than interesting or worthy of further investigation.
